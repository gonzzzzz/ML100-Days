```python
#Create a new function:
def num_missing(x):
  return sum(x.isnull())

#Applying per column:
print("Missing values per column:")
print(app_train.apply(num_missing, axis=0))

#Applying per row:
print("Missing values per row:")
print(app_train.apply(num_missing, axis=1).head())
```

```python
def num_missing0(x):
    return str(x) != 'nan'
app_train['CODE_GENDER'].apply(num_missing0)
```
