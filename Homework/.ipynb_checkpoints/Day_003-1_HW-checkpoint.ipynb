```python
import pandas as pd
import numpy as np
```

### 沒有重複資料

```python
data1 = {'國家':np.random.permutation(["USA","TAIWAN","CHINA","AFRICA","MEXICO"]) ,
        '人口':np.random.randint(12000000,100000001,5) }
data1 = pd.DataFrame(data1)
data1
```

```python
data1=data1.sort_values(["人口"],ascending=False)
data1
```

```python
data1=data1.reset_index(drop=True)
data1
```

```python
data1[0:1]
```

### 重複資料

```python
data2 = {'國家':np.random.permutation(["USA","TAIWAN","CHINA","AFRICA","MEXICO"]*20) ,
        '人口':np.random.randint(12000000,100000001,100) }
data2 = pd.DataFrame(data2)
data2
```

```python
data2=data2.groupby(['國家']).max()
```

```python
data2
```

```python
data2[0:1]
```
